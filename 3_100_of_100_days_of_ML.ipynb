#Step 1: Data Preprocessing


<h4>Importing the libraries</h4>


```python
import pandas as pd
import numpy as np
```


<h4>Importing the dataset</h4>


```python
dataset = pd.read_csv('https://raw.githubusercontent.com/Avik-Jain/100-Days-Of-ML-Code/master/datasets/50_Startups.csv')
X = dataset.iloc[ : , :-1].values
Y = dataset.iloc[ : ,  4 ].values
```


<h4>Encoding Categorical data</h4>


```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import numpy as np

labelencoder = LabelEncoder()
X[:, 3] = labelencoder.fit_transform(X[:, 3])

# Instead of using categorical_features, specify the categories directly
# We assume that the encoded values range from 0 to the maximum encoded value
categories = [np.arange(X[:, 3].max() + 1)]
onehotencoder = OneHotEncoder(categories=categories, sparse_output=False) # sparse=False for dense output

X = onehotencoder.fit_transform(X[:, 3].reshape(-1, 1)) # Reshape for single feature input

# Concatenate the one-hot encoded features with the rest of the data
X = np.concatenate([X[:, :3], X, X[:, 4:]], axis=1)
```


<h4>Avoiding Dummy Variable Trap<h4>


```python
X = X[: , 1:]
```


<h4>Splitting the dataset into the Training set and Test set<h4>


```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```


<h3>Step 2: Fitting Multiple Linear Regression to the Training set</h3>


```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

<h3>Step 3: Predicting the Test set results</h3>

```python
y_pred = regressor.predict(X_test)
```
