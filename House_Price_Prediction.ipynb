```python
#import Lib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#import data
from sklearn.datasets import fetch_california_housing # Use fetch_california_housing instead of load_boston
housing = fetch_california_housing()
```

```python
data= pd.DataFrame(housing.data)
data.columns = housing.feature_names
data.head(10)
```

```python
housing.target.shape
```

```python
data['Price']= housing.target
data.head(10)
```

```python
x=housing.data
y=housing.target
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
print(x_train.shape,x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
#predicting result

y_prediction=regressor.predict(x_test)
```

```python
plt.scatter(y_test,y_prediction,c='red')
plt.plot(y_test,y_test,c='blue')
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Prices vs Predicted Prices")
plt.show
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error
mse=mean_squared_error(y_test,y_prediction)
mae=mean_absolute_percentage_error(y_test,y_prediction)
print("Mean Square Error:",mse)
print("Mean Absolute Error:",mae)
```
